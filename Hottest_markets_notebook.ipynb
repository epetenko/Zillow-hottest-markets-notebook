```python
# CHANGE this for each quarter

peak_quarters = ['2018-Q2', '2018-Q3', '2018-Q4', '2019-Q1']

# For convenience, here are the towns that tend to be in the top that are not county subdivisions.

not_towns = [26765, 37920, 47242]
```

```python
#import our dependancy library
import pandas as pd
```

```python
#will auto-import this quarters' data

df = pd.read_csv('http://files.zillowstatic.com/research/public/City/City_Zhvi_Summary_AllHomes.csv', index_col='RegionID')
```

```python
#lists all values in the state column
df.State.unique()
```

```python
# df.to_csv('Old_data/City_Zhvi_summary_jan19.csv')
```

```python
#filter to NJ data
nj = df[df.State == 'NJ']
```

```python
nj.head()
```

```python
nj.sort_values(by='Zhvi', ascending=False).head()
```

```python
nj.sort_values(by='YoY', ascending=False).head()
```

```python
nj.PeakQuarter.unique()
```

```python
#Filter to cities with peak quarters in past 2 quarters

# filter by that array
njq = nj[nj['PeakQuarter'].isin(peak_quarters)]

njq.head(10)
```

```python
# next sort data by YoY growth

top_nj = njq.sort_values(by='YoY', ascending=False)

top_nj.head(10)
```

```python
top_nj = top_nj[top_nj.index.isin(not_towns) == False]
```

```python
top_nj['LatestRank'] = top_nj['YoY'].rank(ascending=False)
```

```python
# getting previous data
# CHANGE to current file name

old_df = pd.read_csv('Old_data/Parsed_zillow_jan_2019.csv')
```

```python
old_df.columns
```

```python
# You may have to change around the column names to prevent rank name confusion

old_df = old_df.rename(columns={'Rank':'PrevRank', 'RegionName_x':'RegionName'})
```

```python
old_df.head()
```

```python
old_df = old_df[[u'RegionID', u'RegionName', u'PrevRank']]
```

```python
top_nj = top_nj.merge(old_df, left_index=True, right_on='RegionID', how='left')
```
